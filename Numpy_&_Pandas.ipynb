```python
import pandas as pd

# Load the dataset
df = pd.read_csv("retail_dataset .csv")

# Display first 5 rows
print(df.head(5))
```

Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing
values present in the dataset.

```python
# Missing values in each column
print(df.isnull().sum())

# Total missing values in the entire dataset
print("Total Missing Values:", df.isnull().sum().sum())
```

Q3. Replace missing values in the Quantity and Price columns with their respective mean values using
Pandas.

```python
##Replace missing values
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
df['Price'].fillna(df['Price'].mean(), inplace=True)
```

```python
print(df.isnull().sum().sum())
```

Q4. Remove all rows where the Product Category or Region column contains missing values.

```python
##Remove all rows where the Product categpory or region column contains missing values
df.dropna(subset=['Product Category', 'Region'], inplace=True)
```

Q5. Create a new column called Revenue using NumPy, where:

Revenue = Quantity × Price

Display the updated dataset with the new column.

```python
##create new column revenue
import numpy as np
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
print(df.head())
```

Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

```python
## Total Revnue
Total_Revenue = np.sum(df['Revenue'])
print(Total_Revenue)
```

Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.

```python
##Group the dataset by Product Category and calculate the total revenue generated by each category.
Revenue_by_Category = df.groupby('Product Category')['Revenue'].sum()
print(Revenue_by_Category)
```

Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the
highest revenue and the bottom 3 categories with the lowest revenue.

```python
##identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.
Top_3_Categories = Revenue_by_Category.nlargest(3)
Bottom_3_Categories = Revenue_by_Category.nsmallest(3)
```

Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the

```python
Revenue_by_Region = df.groupby('Region')['Revenue'].sum()
print(Revenue_by_Region)
```

Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the
total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column
using NumPy.

```python
import pandas as pd
import numpy as np

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Create Revenue column
df['Revenue'] = df['Quantity'] * df['Price']

# Extract month from Date
df['Month'] = df['Date'].dt.month

# Total monthly revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()
print("Monthly Revenue:")
print(monthly_revenue)

# Mean, Median, Standard Deviation of Revenue
print("Mean Revenue:", np.mean(df['Revenue']))
print("Median Revenue:", np.median(df['Revenue']))
print("Standard Deviation:", np.std(df['Revenue']))
```
